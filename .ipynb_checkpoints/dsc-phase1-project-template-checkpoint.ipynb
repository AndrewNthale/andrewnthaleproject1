```python
# Histogram of "vote_average" column
plt.figure(figsize=(8, 6))
sns.histplot(df['vote_average'], bins=20, kde=True)
plt.title('Distribution of Popularity')
plt.xlabel('vote_average')
plt.ylabel('vote_count')
plt.show()
```

```python
#Countplot for "original_language"
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='original_language', order=df['original_language'].value_counts().index)
plt.title('Distribution of Original Languages')
plt.xlabel('original Language')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Countplot for 'genre_ids'
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='genre_ids', order=df['genre_ids'].value_counts().index)
plt.title('Distribution of Genre IDs')
plt.xlabel('Genre IDs')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
